```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('datasets/tips.csv')
df.head()
```

## Pregunta de investigacion1
## La variable total bill tiene algun comportamiento de alguna distribucion

```python
sns.distplot(df['total_bill'],
             #kde = False,
             bins = 20,
            ).set_title('Histograma de total bill', fontsize= 16)
plt.show()
```

## Conclusion
 La variable 'total_bill' parece 'Weibull Distribution'

```python
sns.boxplot(x = df['total_bill'],
            y = df['day'],
           whis= (0,100), #no mostrar outliers
           ).set_title('Box plot of total_bill')
plt.show()
```
